```python
#BMI CALCULATOR
from tkinter import*
from tkinter import messagebox
root=Tk()
root.geometry("400x300")
root.title("BMI Calculator")

def BMI_cal():
    kg = int(weight_tb.get())
    mtr = int(height_tb.get())/100
    bmi = kg/(mtr*mtr)
    bmi = round(bmi, 1)
    BMI_index(bmi)

def BMI_index(bmi):
    if bmi<18.5:
        messagebox.showinfo("BMI",f"Your BMI={bmi}...You are Underweight")
    elif bmi>18.5 and bmi<24.9:
        messagebox.showinfo("BMI",f"Your BMI={bmi}...You are Normal")
    elif bmi>24.9 and bmi<29.9:
        messagebox.showinfo("BMI",f"Your BMI={bmi}...You are Overwight")
    elif bmi>29.9:
        messagebox.showinfo("BMI",f"Your BMI={bmi}...You are Obese")
    else:
        messagebox.showerror("BMI","Something went wrong!")
        
def reset():
    weight_tb.delete(0,'end')
    height_tb.delete(0,'end')
    age_tb.delete(0,'end')
    
frame1=Frame(root)
frame1.grid(rowspan=5,columnspan=5)
age_lb=Label(frame1,text="Enter your age(1-100)")
age_lb.grid(row=1,column=1)
age_tb=Entry(frame1)
age_tb.grid(row=1,column=2,pady=5)
gender=Label(frame1,text="Select Gender")
gender.grid(row=2,column=1)

frame2=Frame(frame1)
frame2.grid(row=2,column=2,pady=5)
var=IntVar()
rb1=Radiobutton(frame2,variable=var,value=1,text="Male").pack(side=LEFT)
rb2=Radiobutton(frame2,text="Female",variable=var,value=2).pack(side=RIGHT)

weight_lb=Label(frame1,text="Enter your Weight(kg)")
weight_lb.grid(row=3,column=1)
weight_tb=Entry(frame1)
weight_tb.grid(row=3,column=2,pady=5)

height_lb=Label(frame1,text="Enter your Height(cm)")
height_lb.grid(row=4,column=1)
height_tb=Entry(frame1)
height_tb.grid(row=4,column=2,pady=5)

frame3=Frame(frame1)
frame3.grid(row=5,columnspan=3,pady=10)
B1=Button(frame3,text="CALCULATE",command=BMI_cal)
B1.pack(side=LEFT)
B2=Button(frame3,text="RESET",command=reset)
B2.pack(side=LEFT)
B3=Button(frame3,text="EXIT",command=lambda:root.destroy())
B3.pack(side=RIGHT)

root.mainloop()
```
